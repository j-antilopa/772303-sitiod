```python
import numpy as np
```

```python
import matplotlib.pylab as plt
x = np.linspace(-np.pi, np.pi, 201)
plt.plot(x, np.sin(x))
plt.xlabel('Angle [rad]')
plt.ylabel('sin(x)')
plt.axis('tight')
plt.legend(['plot'])
plt.show()
```

```python
left = np.array([[5, 10], [12, 8]]) # Матрица (левая часть системы)
right = np.array([2, 4]) # Вектор (правая часть системы)

np.linalg.solve(left, right)
```
